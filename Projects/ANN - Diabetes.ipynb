# Creating an ANN to predict Diabetes

```python
# Import libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Visualisation libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
from pandas.plotting import scatter_matrix

# Increase the default plot size and set the color scheme
plt.rcParams['figure.figsize'] = 8, 5
plt.style.use("fivethirtyeight")

# Disable warnings 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./data/diabetes.csv')
df.head()
```

```python
sns.pairplot(df, hue = 'Outcome')
```

```python
## Creating numpy arrays for model ##
X = df.drop('Outcome', axis = 1).values
y = df['Outcome']
```

```python
## Splitting the data ##
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Importing all the libraries from PyTorch
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
##### Creating Tensors ######
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train.values)
y_test = torch.LongTensor(y_test.values)
```

```python
df.shape
```

```python
##### Creating model with PyTorch #####
class ANN_Model(nn.Module):
    def __init__(self, input_features = 8, hidden1 = 20, hidden2 = 20, out_features = 2):
        super().__init__()
        self.f_connected1 = nn.Linear(input_features, hidden1)
        self.f_connected2 = nn.Linear(hidden1, hidden2)
        self.out = nn.Linear(hidden2, out_features)
    def forward(self, x):
        x = F.relu(self.f_connected1(x))
        x = F.relu(self.f_connected2(x))
        x = self.out(x)
        return x
```

```python
### Instantiate my ANN_Model
torch.manual_seed(20)
model = ANN_Model()
```

```python
model.parameters
```

```python
### Define the backward Propagation 
## -- Define the loss function 
## -- Define the optimizer

loss_function = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
```

```python
epochs = 500
final_losses = []
for i in range(epochs):
    i = i+1
    y_pred = model.forward(X_train)
    loss = loss_function(y_pred, y_train)
    final_losses.append(loss)
    
    if i % 10 == 1:
        print("Epoch number: {} and the loss: {}".format(i, loss.item()))
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
## Plotting the loss function
plt.plot(range(epochs), final_losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
```

```python
### Prediction in X_test Data
predictions = []
# No gradient with the function torch.no_grad, not needed in testing
with torch.no_grad():
    for i, data in enumerate(X_test):
        y_pred = model(data)
        predictions.append(y_pred.argmax().item())
print(predictions)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predictions)
cm
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(cm, annot = True)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, predictions)
score
```
