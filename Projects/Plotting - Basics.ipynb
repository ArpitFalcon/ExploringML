```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 1. Function method
# 2. Object oriented method
```

### Function Method

```python
x = np.arange(0, 25)
y = x ** 3
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 1000)
y = np.sin(x)
plt.plot(x, y)

# Cos x
x = np.linspace(0, 4*np.pi, 1000)
y = np.cos(x)
plt.plot(x, y, color = 'red')
plt.xlabel('X')
plt.ylabel('Sin(x) and Cos(x)')
plt.title('Different functions')
plt.show()
```

```python
# Subplot
# plt.subplot(number of rows, number of columns, plot number where the figure is desired)

plt.subplot(1, 2, 1)
plt.plot(x, y, color = 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, color = 'b')
plt.show()
```

```python
plt.subplot(221)
plt.plot(x, y, color = 'r')

plt.subplot(222)
plt.plot(x, y, color = 'b')

plt.subplot(223)
plt.plot(x, y, color = 'g')

plt.subplot(224)
plt.plot(x, y, color = 'purple')

plt.tight_layout() # Used when the graphs is overlapping with each other
```

### Object Oriented Method

```python
fig = plt.figure()
# axes = fig.add_axes([bottom x, bottom y, width, height])  #Empty canvas
axes1 = fig.add_axes([0.3, 0.2, 0.8, 0.8])
axes1.plot(x, y)
axes1.set_xlabel('x1_Params')
axes1.set_ylabel('y1_Params')
axes1.set_title('Bada Graph')


axes2 = fig.add_axes([0, 0.4, 0.4, 0.4])
axes2.plot(x, y)
axes2.set_xlabel('x2_Params')
axes2.set_ylabel('x2_Params')
axes2.set_title('Chota Graph')


plt.show()
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3, figsize = (10, 6))
axes[0][0].plot(x, y)
plt.tight_layout()
# fig.savefig('graph.png')
```

### Distribution Plots

```python
# Loading dataset from seaborn library
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
# 1. Dist Plot - Univariate
sns.distplot(tips['total_bill'], bins = 10) # KDE - Kernel Distributions Estimation 
```

```python
# 2. Joint Plot - Bivariate
sns.jointplot(data = tips, x = 'tip', y = 'total_bill', kind = 'reg', color = 'purple')
```

```python
# 3. Pair Plot - Multivariate
sns.pairplot(tips, hue='smoker', palette = 'magma')
```

```python
# 4. Rug Plot - 
sns.rugplot()
```
