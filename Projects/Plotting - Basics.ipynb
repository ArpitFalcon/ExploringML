# Matplot Lib Basics

```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 1. Function method
# 2. Object oriented method
```

### Function Method

```python
x = np.arange(0, 25)
y = x ** 3
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 1000)
y = np.sin(x)
plt.plot(x, y)

# Cos x
x = np.linspace(0, 4*np.pi, 1000)
y = np.cos(x)
plt.plot(x, y, color = 'red')
plt.xlabel('X')
plt.ylabel('Sin(x) and Cos(x)')
plt.title('Different functions')
plt.show()
```

```python
# Subplot
# plt.subplot(number of rows, number of columns, plot number where the figure is desired)

plt.subplot(1, 2, 1)
plt.plot(x, y, color = 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, color = 'b')
plt.show()
```

```python
plt.subplot(221)
plt.plot(x, y, color = 'r')

plt.subplot(222)
plt.plot(x, y, color = 'b')

plt.subplot(223)
plt.plot(x, y, color = 'g')

plt.subplot(224)
plt.plot(x, y, color = 'purple')

plt.tight_layout() # Used when the graphs is overlapping with each other
```
