```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from keras import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def create_model(cols):
    model = Sequential()
    
    model.add(Dense(12, activation = 'relu', input_shape=(cols,)))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    return model
```

```python
seed = 7
np.random.seed(seed)
```

```python
data_train = np.loadtxt('./data/diabetes_data.csv', skiprows=1, delimiter=',')
```

```python
data_train
```

```python
X = data_train[:, 0:8]
y = data_train[:, 8]
```

## 5 Creating a Keras Model using the sklearn wrapper class

```python
model = KerasClassifier(build_fn = create_model, verbose = 1, cols = 8)
```

## 6 Defining the grid search elements

```python
batch_size = [10, 20, 40, 60, 80, 100]
epochs = [10, 50, 100]
```

```python
param_grid = dict(batch_size=batch_size, epochs=epochs)
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
```

```python
param_grid
```

## 7 Passing the whole grid to the model for training

```python
grid_result = grid.fit(X, y)
```

## 8 Summarizing the results

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

pd.DataFrame(zip(means, stds, params), columns=['mean', 'std dev', 'parameters'])
```
