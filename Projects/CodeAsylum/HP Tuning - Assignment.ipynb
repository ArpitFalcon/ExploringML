## Importing Libraries 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from keras import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import to_categorical
from keras.wrappers.scikit_learn import KerasClassifier
```

## Build Model Class

```python
def create_model(cols, optimizer='adam', activation='relu', dropout=0.2):
    model = Sequential()
    
    model.add(Dense(12, activation=activation, input_shape=(cols,)))
    model.add(Dropout(dropout))
    model.add(Dense(6, activation=activation))
    model.add(Dropout(dropout))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    
    return model
```

```python
seed = 7
np.random.seed(seed)
```

```python
data_train = np.loadtxt('./data/diabetes_data.csv', skiprows=1, delimiter=',')
```

```python
X = data_train[:, 0:8]
y = data_train[:, 8]
```

## 5 Creating a Keras Model using the sklearn wrapper class

```python
modelToTune = KerasClassifier(build_fn = create_model, verbose = 1, cols = 8)
```

## Grid Search Elements

```python
batch_size = [10, 50, 100]
epochs = [10, 15, 20]
optimizer = ['Adam', 'Adadelta', 'Adamax', 'Adagrad', 'SGD', 'Nadam']
dropout = [0.1, 0.2, 0.3]
activation = ['relu', 'sigmoid', 'tanh']
```

```python
param_grid = dict(batch_size=batch_size, epochs=epochs, optimizer=optimizer, dropout=dropout, activation=activation)
grid = GridSearchCV(estimator=modelToTune, param_grid=param_grid, cv=3)
```

```python
param_grid
```

##  Passing the whole grid for model training

```python
grid_result = grid.fit(X, y)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
